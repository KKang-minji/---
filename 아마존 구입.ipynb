```python
import pandas as pd # DataFrame 을 위한 모듈
import numpy as np # arrange 
import matplotlib.pyplot as plt # 시각화 모듈
```

```python
# ,와 ₹ 제거
order['ship_city']=order['ship_city'].apply(lambda x: x.replace(',',''))
order["ship_state"] = order["ship_state"].str.replace(",", "")
order["shipping_fee"] = order["shipping_fee"].str.replace("₹", "")
order["item_total"] = order["item_total"].str.replace("₹", "")
```

```python
order.isnull().sum()
```

```python
order.head()
```

```python
order['shipping_fee'].fillna(order['shipping_fee'].mode()[0], inplace=True)

order['item_total'].fillna(order['item_total'].mode()[0], inplace=True)

order['cod'].fillna('online', inplace=True)
```
